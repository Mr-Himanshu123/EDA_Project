````python
#Manupulating 'Size' into Numerical column with removal of 'k','M','+' from Dataset
def size_convert(app_size):
  '''
  This is the function to convert the size into kb into mb
  '''
  try:
    if app_size[-1]=='M':
      return float(app_size[:-1])
    elif app_size[-1]=='k':
      return round((float(app_size[:-1])/1024),2)
    else:
      return float(app_size)
  except:
    return app_size
````

````python
#convert the column as suggested
playstore_df['Size']=playstore_df['Size'].apply(lambda size: size_convert(size))
#check the datatype of column after applying function on the column
playstore_df['Size'].dtype
````

**(C) Merging of Two DataFrame**

````python
#create a new dataset by grouping App And Sentiment.
gb_df=userReview_df.groupby(['App','Sentiment']).size().unstack(level=1)
gb_df.reset_index(inplace=True)
gb_names=['App','Negetive_Sentiment','Neutral_Sentiment','Positive_Sentiment']
gb_df.columns=gb_names
gb_df.head()
````

````python
#Add mean sentiment polarity and mean sentiment to the dataframe
app_list=list(userReview_df['App'].unique())
sen_pol=[]
sen_sub=[]
for i in app_list:
  app_df=userReview_df[userReview_df['App']==i]
  sen_pol.append(app_df['Sentiment_Polarity'].mean())
  sen_sub.append(app_df['Sentiment_Subjectivity'].mean())
gb_df['Mean_Sentiment_polarity']=sen_pol
gb_df['Mean_Sentiment_Subjectivity']=sen_sub
gb_df.head()
````

````python
#Find null values in new gb_df
gb_df.isnull().sum()
````

````python
#Replacing null values with Zero
gb_df.fillna(0,inplace=True)
````

````python
#Merging Two Dataset with respect to inner joint
merged_df=pd.merge(playstore_df,gb_df, on='App', how='inner')
merged_df.head(6)
````

### **4.1.6 Analysis of Rating Column**

to get more clear understanding about rating column dataframe can be convert into categorical data like below.

````python
#create a group rating function
def app_rate(rating):
  '''
  This function helps to categorize the app rating
  '''
  if rating>=4.0:
    return 'Top Rated'
  elif rating>=3.0 and rating<4.0:
    return 'Above Average'
  elif rating>=2.0 and rating<3.0:
    return 'Average'
  else:
    return 'below average'
````

````python
#Create a column in the dataframe and apply the function in the column
playstore_df['Rating Group']=playstore_df['Rating'].apply(lambda x:app_rate(x))
playstore_df.head(4)
````

### **4.2.4 Installs Vs App size**

For better analysis, we have to create an app-sized group.

````python
#create app_size group
def app_size_group(size):
  '''
  This function is creating a group of appsize between 0-100MB
  '''
  try:
    if size<1.0:
      return 'Below 1MB'
    elif size>=1.0 and size<5.0:
      return 'Below 5MB'
    elif size>=5.0 and size>15.0:
      return'5-15MB'
    elif size>=15.0 and size<30.0:
      return '15-30MB'
    elif size>=30.0 and size<60.0:
      return '30-60MB'
    elif size>=60.0 and size<100.0:
      return '60-100MB'
    else:
      return 'Above 100MB'
  except:
    return size
````

````python
#Apply the function to the dateframe
playstore_df['appsize_category']=playstore_df['Size'].apply(lambda x:app_size_group(x))
playstore_df.head()
````

**1) Tools and Medical apps, Is the average rating significantly different in each category?**

````python
#List out the rating data of two category
tools =list(playstore_df[playstore_df.Category=='TOOLS'].Rating)
medical=list(playstore_df[playstore_df.Category=='MEDICAL'].Rating)
#check if the plots are normal
sns.kdeplot(tools)
sns.kdeplot(medical)
plt.title('Rating distributions')
plt.show()
````

````python
#make numpy array of the data
tools_array=np.asarray(tools)
medical_array=np.asarray(medical)
#calculate the standard deviation
print('Standard deviation of tools app rating:',tools_array.std())
print("Standard deviation of medical app ratings:",medical_array.std())
````

````python
#import statistics library to run the hypothesis testing
from scipy.stats import ttest_ind
#confidence interval: 95%
#setting confidence interval sets our alpha (treshold value)=1-0.95=0.05

#Null Hypothesis : Difference in the mean rating of Tools and Medical apps are due to a random chance
#Alternative Hypothesis:Mean rating of Tools and Medical apps are significantly different

#p-value -when it is assumend that our null hypothesis is correct, p value gives us the probability of
#getting a sample with the result we assumed.

#run the 2 sample test:

_,pvalue=ttest_ind(tools,medical)
if pvalue<=0.05:
  print('Reject Null Hypothesis')
else:
  print('Accept Null hypothesis')
````
